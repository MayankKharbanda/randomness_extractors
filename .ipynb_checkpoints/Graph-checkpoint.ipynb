```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
input_size = [50, 100, 500, 1000, 6000]
universal_time = [703.8618054389954, 1407.0338921546936, 7023.768018245697, 14081.416084766388, 86225]
universal_size = [37500000, 75000000, 375000000, 750000000, 4500000000]

neumann_time = [178.0420858860016, 356.1498944759369, 1779.9884378910065, 3560.330964565277, 17973]
neumann_size = [12499757, 25000037, 125001854, 250005463, 1499987850]

toeplitz_time = [8026.742326974869, 16077.618845462799, 80372.7703871727, 159904.2883388996]
toeplitz_size = [37500000, 75000000, 375000000, 750000000]
```

```python
fig = plt.figure(figsize=(7,7))
plt.plot(input_size,np.divide(universal_size,universal_time), marker='o')
plt.plot(input_size,np.divide(neumann_size, neumann_time), marker='o')
plt.plot(input_size[:-1],np.divide(toeplitz_size, toeplitz_time), marker='o')
plt.xlabel('file_size(MB)')
plt.ylabel('Bytes/s')
fig.legend(['universal','neumann','toeplitz'])
plt.show()
fig.savefig('speed.png', bbox_inches='tight')
```
