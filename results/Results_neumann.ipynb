```python
location_result = 'results/neumann'
no_files = 2
```

```python
small_crush = [0]*11
for fl in range(1,no_files+1):
    for te in range(1,11):
        with open(f'{location_result}/small_crush_result/file_{fl}/small_crush/{te}.txt','r') as fr:
            lines = fr.readlines()
            temp=1
            while(True):
                if(lines[-temp].strip() == ''):
                    temp+=1
                    continue
                elif 'passed' in lines[-temp]:
                    small_crush[te] = small_crush[te] +1
                break
```

```python
crush = [0]*97
for fl in range(1,no_files+1):
    for te in range(1,97):
        with open(f'{location_result}/crush_result/file_{fl}/crush/{te}.txt','r') as fr:
            lines = fr.readlines()
            temp=1
            while(True):
                if(lines[-temp].strip() == ''):
                    temp+=1
                    continue
                elif 'passed' in lines[-temp]:
                    crush[te] = crush[te] +1
                break
```

```python
rabbit = [0]*27
for fl in range(1,no_files+1):
    for te in range(1,27):
        with open(f'{location_result}/rabbit_result/file_{fl}/rabbit/{te}.txt','r') as fr:
            lines = fr.readlines()
            temp=1
            while(True):
                if(lines[-temp].strip() == ''):
                    temp+=1
                    continue
                elif 'passed' in lines[-temp]:
                    rabbit[te] = rabbit[te] +1
                break
```

```python
alphabit = [0]*10
for fl in range(1,no_files+1):
    for te in range(1,10):
        with open(f'{location_result}/alphabit_result/file_{fl}/alphabit/{te}.txt','r') as fr:
            lines = fr.readlines()
            temp=1
            while(True):
                if(lines[-temp].strip() == ''):
                    temp+=1
                    continue
                elif 'passed' in lines[-temp]:
                    alphabit[te] = alphabit[te] +1
                break
```

```python
nist = [0]*16
FLAG = 0
for i in range(1, no_files+1):
    for j in range(1,16):
        with open(f'{location_result}/nist_result/file_{i}/nist/{j}/finalAnalysisReport.txt','r') as fr:
            lines = fr.readlines()
            FLAG=0
            for line in lines[7:]:
                if (line.strip()==''):
                    break
                data=line.split()
                if(data[-2].find('/')==-1):
                    p_val=(int(data[-3].split('/')[0]))    
                else:
                    p_val=(int(data[-2].split('/')[0]))
                if(j==12 or j==13): 
                    if(p_val<85):
                        FLAG=1
                        break
                else:
                    if(p_val<96):
                        FLAG=1
                        break
            if(FLAG==0):
                nist[j]+=1
```

```python
ent_test = {
    #'File_bytes' : 0,
    'Entropy' : 1,
    'Chi-square' : 2,
    'Mean' : 3,
    'Monte-Carlo' : 4,
    'Corr' : 5
    }
ent_stat = {
    'Count' : 0,
    'Fract' : 1
    }
```

```python
result_arr = [None]*no_files
result_stat_arr = []
for i in range(no_files):
    result_stat_arr.append([None]*256)
for i in range(no_files):
    with open(f'{location_result}/ent_result/{i}.txt','r') as fr:
        lines = fr.readlines()
        result_arr[i] = [float(x.strip()) for x in lines[1].split(',')[1:]]
        number = 0
        for line in lines[3:]:
            result_stat_arr[i][number] = [float(x.strip()) for x in line.split(',')[2:]]
            number = number+1
```

```python
fig,ax =  plt.subplots(figsize=(12,15), sharex=True)
fig.tight_layout(pad=0.3)
fig.text(0.5, 0.0, 'Tests_iteration', ha='center')
fig.text(-0.05, 0.5, 'Value', va='center', rotation='vertical')

v_line = [None, 8, 256, 127.5, 3.14159, 0]


for test,value in ent_test.items():
    
    plt.subplot(3,2,value)
    plt.scatter(range(1, no_files+1),[result[value] for result in result_arr])
    plt.plot([1,no_files+1], [v_line[value], v_line[value]], color = 'g')
    if(value == 2):
        plt.ylim(250,570)
    plt.title(test)
    

plt.show()
fig.savefig('fig/ent.png', bbox_inches='tight')
```
